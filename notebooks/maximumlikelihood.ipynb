```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from matplotlib import pyplot as plt
```

```python
np.random.seed(seed=233423)

#Generate predetermined regressor.
n=1000
X=np.linspace(1,100,n)[:, np.newaxis]
#Add constant
X = np.concatenate((np.ones(n)[:, np.newaxis],X),axis=1)

#Set constant and regressor coefficients
beta=np.array([3,1])[:,np.newaxis]
sigma = 2
u=stats.norm.rvs(0,sigma,(n,1))

y = X.dot(beta)+u

plt.plot(y)
plt.show()
```

```python
XTX = np.dot(X.T,X)
beta_hat = np.linalg.multi_dot([np.linalg.inv(XTX),X.T,y])

y_hat = X.dot(beta_hat)
u_hat = y-y_hat
s2 = (1/(n-X.shape[1]))*sum(np.square(u_hat))
beta_hat_cov = s2*np.linalg.inv(XTX)

t = beta_hat[0]/beta_hat_cov[1,1]
```
